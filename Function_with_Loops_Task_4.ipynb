```python
# Problem 1: Mobile Recharge Validation System
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    # Rule 1: Minimum recharge amount
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False

    # Rule 2: Check if plan is valid
    if amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans: 199, 299, 399, 599")
        return False

    return True


# Retry loop
while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))

        if validate_recharge(amount):
            print(" Recharge Successful!")
            break
        else:
            print(" Please try again.\n")

    except ValueError:
        print("Please enter a valid numeric amount.\n")
```

```python
# Problem 2: Inventory Reorder Alert System
def check_stock(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK (Stock = {quantity})")


# Dictionary of products and their stock
inventory = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 12,
    "Soap": 30,
    "Shampoo": 8
}

# Function Call
check_stock(inventory)
```

```python
#  Problem 3: Student Result Processing System
def evaluate_result(marks):
    total = 0

    # Loop to calculate total
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Average Marks: {average:.2f} → Pass"
    else:
        return f"Average Marks: {average:.2f} → Fail"


student_marks = [60, 45, 70, 55, 40]

# Function Call
result = evaluate_result(student_marks)
print(result)
```

```python
# Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare(distance, peak):
    base_fare = 50
    rate_per_km = 12

    fare = base_fare + (distance * rate_per_km)

    if peak:
        fare = fare * 1.25   # 25% extra during peak hours

    return fare


# Retry loop
while True:
    try:
        distance = float(input("Enter distance travelled (in km): "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()

        peak = True if peak_input == "yes" else False

        total_fare = calculate_fare(distance, peak)

        print(f"Total Fare: ₹{total_fare:.2f}")
        break

    except ValueError:
        print("Invalid input. Please enter valid numbers.\n")
```

```python
# Problem 5: Employee Attendance Eligibility Checker
def check_eligibility(attendance):
    present_count = 0

    for day in attendance:
        if day == "P":
            present_count += 1

    percentage = (present_count / len(attendance)) * 100

    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% → Eligible"
    else:
        return f"Attendance: {percentage:.2f}% → Not Eligible"


attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]

# Function call
result = check_eligibility(attendance_list)
print(result)
```

```python
# Problem 6: Password Strength Checker
def check_password_strength(password):
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character @#$)"


user_password = input("Enter password: ")
result = check_password_strength(user_password)
print(result)
```
