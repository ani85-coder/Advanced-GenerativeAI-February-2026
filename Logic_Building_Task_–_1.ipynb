```python
#Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
pass_count=0
fail_count=0
for i in marks:
  if i>=50:
    pass_count += 1
  else:
    fail_count += 1

print('Total Students',len(marks))
print('Total number of passed students : ',pass_count)
print('Total number of failed students : ',fail_count)
```

```python
# Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
updated_names=[x.strip().lower() for x in names]
print('Updated Names : ',updated_names)
```

```python
# Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
  length=len(i)
  print(f"Messages : '{i}' | Length :{length}",end ="")

  if length > 10:
    print("  --> FLAG: Longer than 10 characters")
  else:
    print()
```

```python
# Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count=0
for i in logs:
  if i == 'ERROR':
    error_count += 1
  else:
    error_count += 0

print('Number of ERROR entries : ',error_count)
```
