```python
# 1.Social Media Post Engagement Analyzer
def post_engagement_analyzer(likes_list):
    total_likes = 0

    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


# Example
likes = [200, 300, 250, 300]
total, result = post_engagement_analyzer(likes)

print("Total Likes:", total)
print("Post Status:", result)
```

```python
# 2. Healthcare  medical Stock Alert System
def medicine_stock_alert(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    return stock, status


stock, result = medicine_stock_alert(6)

print("Medicine Stock:", stock)
print("Status:", result)
```

```python
# 3.Agriculture Rainfall Adequacy Checker
def rainfall_checker(rainfall_data, required_level):
    total = 0

    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average, status


rainfall = [70, 75, 68, 80]
avg, result = rainfall_checker(rainfall, 72)

print("Average Rainfall:", avg)
print("Rainfall Status:", result)
```

```python
# 4.Duplicate Account Detection
def duplicate_account_detector(usernames):
    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        return "Duplicate Accounts Found: Yes"
    else:
        return "Duplicate Accounts Found: No"


users = ["omkar", "rahul", "omkar"]
print(duplicate_account_detector(users))
```

```python
# 5.Appointment Eligibility Checker
def appointment_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    return age, status


age, result = appointment_eligibility(21)

print("Patient Age:", age)
print("Eligibility Status:", result)
```

```python
# 6.Premium Crop Price Filter
def premium_crop_filter(price_list):
    premium = []

    for price in price_list:
        if price > 2000:
            premium.append(price)

    return premium


prices = [1500, 2500, 3200, 1800]
print("Premium Crops:", premium_crop_filter(prices))
```

```python
# 7.Application Health Checker
def app_health_checker(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    return error_count, status


errors, result = app_health_checker(7)

print("Error Count:", errors)
print("System Status:", result)
```

```python
# 8.Transaction Limit Checker
def transaction_checker(amount):
    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    return amount, status


amount, result = transaction_checker(60000)

print("Transaction Amount:", amount)
print("Transaction Status:", result)
```

```python
# 9.Attendance Eligibility System
def attendance_checker(attendance_list):
    total_classes = len(attendance_list)
    attended = sum(attendance_list)

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    return percentage, status


attendance = [1,1,1,0,1,1,1,1,0,1]
percent, result = attendance_checker(attendance)

print("Attendance Percentage:", percent)
print("Exam Eligibility:", result)
```

```python
# 10.Smart Electricity Bill Analyzer
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Classification
    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


bill_amount, usage_status = electricity_bill(250)

print("Total Bill:", bill_amount)
print("Usage Status:", usage_status)
```
