# 1:Smart Parking Lot Management System

```python

def smart_parking(capacity, logs):
    current_parked = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            current_parked -= 1

        # Track peak usage
        if current_parked > peak_usage:
            peak_usage = current_parked

    print("Currently Parked Vehicles:", current_parked)
    print("Peak Parking Usage:", peak_usage)

    # Capacity Alert
    if current_parked > capacity:
        print(" ALERT: Parking capacity exceeded!")
    else:
        print("Parking Status: Available")

capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, logs)
```

2:Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):

    # Base speed (km per hour)
    base_speed = 30

    # Base time calculation (convert hours to minutes)
    base_time = (distance / base_speed) * 60

    # Traffic adjustments
    if traffic == "Low":
        base_time += 5
    elif traffic == "Medium":
        base_time += 10
    elif traffic == "High":
        base_time += 22

    # Weather adjustments
    if weather == "Rainy":
        base_time += 17
    elif weather == "Storm":
        base_time += 25

    print("Estimated Delivery Time:", round(base_time), "minutes")

estimate_delivery_time(8, "High", "Rainy")
```

3:Movie Theatre Seat Occupancy

```python
def analyze_show(total_seats, booked_seats):

    # Count booked seats
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print("\nOccupancy:", round(occupancy), "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# ---- Input Section ----

total_seats = int(input("Enter Total Seats: "))

booked_number = int(input("Enter Number of Booked Seats: "))

# Creating list dynamically
booked_seats = []
for i in range(booked_number):
    booked_seats.append(1)

# Call function
analyze_show(total_seats, booked_seats)
```

4:Cloud Server Load Classification System


```python
def classify_server_load(cpu_readings):

    total = 0

    # Calculate total using loop
    for reading in cpu_readings:
        total += reading

    # Calculate average
    average = total / len(cpu_readings)

    print("\nAverage CPU Load:", round(average), "%")

    # Classification
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


#  Input Section

n = int(input("Enter number of CPU readings: "))

cpu_readings = []

for i in range(n):
    value = float(input(f"Enter CPU reading {i+1} (%): "))
    cpu_readings.append(value)

# Call function
classify_server_load(cpu_readings)
```

5:Smart Classroom Resource Usage Monitor

```python
def track_resource_usage(resource_usage):

    overused = []
    threshold = 8   # Define overuse limit (hours)

    # Loop through dictionary
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)

    # Output Results
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# -------- Sample Input --------

resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

track_resource_usage(resource_usage)
```

6: Online Event Registration Capacity Controller

```python
def manage_event_registrations(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0

    # Loop through each registration request
    for _ in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1  # waitlist mode after capacity is full

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if confirmed == capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


#  Input
capacity = int(input("Enter Event Capacity: "))
total_registrations = int(input("Enter Total Registrations: "))

manage_event_registrations(capacity, total_registrations)
```
