```python
# 1.Employee Performance Bonus Eligibility
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest_score = max(employees.values())

top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", highest_score, ")")
```

```python
# 2: Search Query Keyword Analysis
query = "Buy mobile phone buy phone online"

query = query.lower()

cleaned = ""
for ch in query:
    if ch.isalnum() or ch == " ":
        cleaned = cleaned + ch

words = cleaned.split()

frequency = {}

for word in words:
    if word in frequency:
        frequency[word] = frequency[word] + 1
    else:
        frequency[word] = 1

result = {}

for word in frequency:
    if frequency[word] > 1:
        result[word] = frequency[word]

print(result)
```

```python
# 3: Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid_readings.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# 4: Email Domain Usage Analysis
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

total_emails = len(emails)

for email in emails:
    parts = email.split("@")
    domain = parts[1]

    if domain in domain_count:
        domain_count[domain] = domain_count[domain] + 1
    else:
        domain_count[domain] = 1

for domain in domain_count:
    percentage = (domain_count[domain] * 100) / total_emails
    print(domain + ":", int(percentage), "%", sep="")
```

```python
# 5: Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]

total = 0

for value in sales:
    total = total + value

average = total / len(sales)

threshold = average + (average * 0.30)

for i in range(len(sales)):
    if sales[i] > threshold:
        print("Day", i + 1, ":", sales[i])
```

```python
# 6: Duplicate User ID Detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count_dict = {}

for user in user_ids:
    if user in count_dict:
        count_dict[user] = count_dict[user] + 1
    else:
        count_dict[user] = 1

for user in count_dict:
    if count_dict[user] > 1:
        print(user, "→", count_dict[user], "times")

```
