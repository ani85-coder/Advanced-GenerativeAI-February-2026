1: Smart Light Controller


```python
class SmartLight:
    def __init__(self, name: str):
        self.name = name
        self.is_on = False  # default state

    def turn_on(self):
        self.is_on = True

    def turn_off(self):
        self.is_on = False

    def status(self) -> str:
        return f"{self.name} is {'ON' if self.is_on else 'OFF'}"



light_name = "Bedroom Light"
action = "ON"

light = SmartLight(light_name)

if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()

print(light.status())
```

2: Employee ID Card System

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

employee_name = "Rahul"
employee_id = "EMP102"
department = "IT"

emp = Employee(employee_name, employee_id, department)
emp.display_id_card()
```

3: Mobile Contact Record


```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

contact_name = "Anita"
phone_number = "9876543210"

contact = Contact(contact_name, phone_number)
contact.display_contact()
```

4: Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

product_name = "Headphones"
price = 2499

product = Product(product_name, price)
product.print_price_tag()
```

 5: Movie Rating Display System

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

movie_name = "Inception"
rating = 4.8

movie = Movie(movie_name, rating)
movie.display_details()
```

6: Delivery Address Manager

```python
class Customer:
    def __init__(self, name, address):
        self.name = name
        self.address = address

    def print_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.name}")
        print(f"Address: {self.address}")

customer_name = "Suman"
address = "Hyderabad"

customer = Customer(customer_name, address)
customer.print_delivery_details()
```
